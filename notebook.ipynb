```python
%%shell

for animal in c.elegans ciliate drosophila e.coli human methanocaldococcus mouse thermococcus tuberculosis yeast zebrafish
do
   for protein in H2A_b H2B_b H3_b H4_b KMT2C_b
   do
   	echo "$animal $protein";
   	blastp -query /content/data/histones/$protein.fasta -db /content/data/proteomes/$animal.faa -out results/$protein-$animal.blast -outfmt 7;
   done
done
```

```python
import pandas as pd
import numpy as np

proteomes = "c.elegans ciliate drosophila e.coli human methanocaldococcus mouse thermococcus tuberculosis yeast zebrafish".split()
proteins = "H2A_b H2B_b H3_b H4_b KMT2C_b".split()

data = dict()

for proteome in proteomes:
  data[proteome] = dict()
  for protein in proteins:
    with open(f"results/{protein}-{proteome}.blast", 'r') as f:
      res = float(f.readlines()[5].split()[10])
    data[proteome][protein.replace('_a', '') if '_a' in protein else protein] = max(0, -np.log10(res if res != 0 else 1e-300))


df = pd.DataFrame.from_records(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = df.reindex(["human", "mouse", "zebrafish", "drosophila", "c.elegans", "ciliate", "yeast", "methanocaldococcus", "thermococcus", "e.coli", "tuberculosis"], axis=1)
plt.figure(figsize=(10, 8))
sns.heatmap(df, cmap='coolwarm')

for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        plt.text(j + 0.5, i + 0.5, f'{df.iloc[i, j]:.2f}', ha='center', va='center')

plt.xlabel('Протеомы')
plt.ylabel('Белки')

plt.title('Тепловая карта')

plt.show()
```
